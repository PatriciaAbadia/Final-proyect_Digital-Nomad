### * Diccionario entre nombre de las columnas y correspondiente valor por ciudad. Se llama 'info_ciudades'

```python
urlbase = 'https://nomadlist.com/'
info_ciudades = []
for i in cities:
    driver.get(urlbase + i)
    score=['Ciudad']
    score_desc=[f'{i}']
    cont=1
    while True:
        try:
            score.append(driver.find_element_by_css_selector(f'body > div.modal.item.show.loaded > ajaxloaded > div:nth-child(7) > div.expandBelow > div.tab-scroller-container > div > div.tab.tab-ranking.show > table > tbody > tr:nth-child({cont}) > td.key').text)
            score_desc.append(driver.find_element_by_css_selector(f'body > div.modal.item.show.loaded > ajaxloaded > div:nth-child(7) > div.expandBelow > div.tab-scroller-container > div > div.tab.tab-ranking.show > table > tbody > tr:nth-child({cont}) > td.value > div > div').text)
            cont +=1
        except:
            break
    dic_ciudad = dict(zip(score,score_desc))
    info_ciudades.append(dic_ciudad)
```
